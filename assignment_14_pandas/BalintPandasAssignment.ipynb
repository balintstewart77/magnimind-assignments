11- Merge these two DataFrames into a single DataFrame called census.

```python
# Load the two datasets to be merged
populations = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')
divisions = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')
```

```python
# Preview the data to identify column to merge on
populations.head()
```

```python
# Preview the data to identify column to merge on
divisions.head()
```

```python
# Merge the datasets on 'state'
census = populations.merge(divisions, on = 'state')

census.head()
```

12- Re-shape census so that one column contains all population measures, and another the year attributes.


```python
# Reshape the census data into 'long' format using pd.melt() function
census_long = pd.melt(census, id_vars = ['state','division','region'], var_name = 'Year', value_name = 'Population')
census_long.head()
```

13- Group the data by year and summarize it.

```python
# Convert 'Population' column to numeric type in order to carry out agg function
census_long['Population'] = pd.to_numeric(census_long['Population'])
```

```python
# Summarise the data using .agg() function and some standard statistical summaries
summary = census_long.groupby('Year').agg(total_pop = ('Population', 'sum'),
                                mean_pop = ('Population','mean'),
                                std_pop = ('Population', 'std'),
                                max_pop = ('Population', 'max'),
                                min_pop = ('Population','min'))
print(summary)
```

14- Group the data by region, division and year and summarize it.

```python
summary2 = census_long.groupby(['region', 'division','Year']).agg(total_pop = ('Population', 'sum'),
                                mean_pop = ('Population','mean'),
                                std_pop = ('Population', 'std'),
                                max_pop = ('Population', 'max'),
                                min_pop = ('Population','min'))
print(summary2)
```
