# B. Loading a dataset and exploring it

```python
# Import the modules needed to explore the data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#  Import auto_mpg.csv dataset using pandas'read_csv function. NB THIS IS FOR GOOGLE COLAB ONLY!!
# URL of the CSV file
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original"

# Column names based on the dataset description
column_names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]

# Read the CSV file from the URL
df = pd.read_csv(url, delim_whitespace=True, names=column_names)

# Print the first three samples from your dataset
df.head(3)
```

```python
# Print the index range of the observations
df.index

# Print the column names of your dataset
df.columns
```

```python
# Assign mpg column as output and name it as y and the rest of the data as the features and assign it to X. Print the shape of X.
y = df['mpg']
X = df.drop('mpg', axis=1)
print(X.shape)
```

```python
# BONUS: Visualize the relationships between these data points.
# Another way to look at relationships between the variables is to use heatmap of correlations between the variables
# Need to drop categorical varaibles first
cat_vars = ['model_year', 'origin', 'car_name']
Xb = X.drop(cat_vars, axis = 1)
corr = Xb.corr()
#Draw heatmap of correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
def scale_feature(data):
    min_val = data.min()
    max_val = data.max()
    scaled_data = (data - min_val / (max_val-min_val))
    return scaled_data

# Scaling features
df['displacement_scaled'] = scale_feature(df['displacement'])
df['horsepower_scaled'] = scale_feature(df['horsepower'])
df['acceleration_scaled'] = scale_feature(df['acceleration'])
df['weight_scaled'] = scale_feature(df['weight'])
df['mpg_scaled'] = scale_feature(df['mpg'])
```

```python
# Create a boxplot of mpg for different origin values before and after scaling.
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(x ='origin', y ='mpg', data = df, ax = axes[0])
axes[0].set_title('Before Scaling')
sns.boxplot(x = 'origin', y = 'mpg_scaled', data = df, ax = axes[1])
axes[1].set_title('After Scaling')
plt.show()
```
